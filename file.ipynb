```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data preparation
data = {
    'Variable Subgroup': [
        'Asia', 'Africa', 'America', 'Multiple Countries',
        'High Income Country', 'Middle Income Country', 'Low Income Country', 'Multiple Countries/Regions',
        'Before 2020', '2020-2049', '2050-2100',
        'Cross sectional (before 2020)', 'Prospective Cohort (2020-2100)',
        'Drought', 'Flood', 'RCP scenario', 'Precipitation-Temperature',
        'Increase Production', 'Decrease Production', 'Not Report',
        'Decrease GDP', 'Increase GDP', 'Non report GDP',
        'Medium Quality', 'High quality'
    ],
    'Effect Size (mean, 95% CI)': [
        (23, 13, 65), (28, 9, 52), (14, 6, 25), (20, 6, 39),
        (23, 6, 47), (34, 4, 75), (35, 19, 53), (13, 5, 24),
        (28, 4, 63), (31, 8, 62), (22, 9, 38),
        (35, 5, 76), (23, 12, 37),
        (32, 15, 52), (8, 6, 10), (17, 6, 30), (36, 16, 59),
        (20, 19, 21), (31, 15, 49), (10, 9, 11),
        (28, 12, 47), (31, 5, 66), (24, 6, 48),
        (27, 8, 52), (25, 10, 44)
    ],
    'I2 (%)': [
        99.84, 99.47, 0.0, 99.73,
        99.69, 99.93, 99.49, 98.65,
        99.85, 99.89, 99.63,
        99.88, 99.78,
        99.44, 0.0, 99.64, 99.77,
        0.00, 99.84, 0.00,
        99.40, 99.83, 99.90,
        99.88, 99.80
    ]
}
```

```python
# Create DataFrame
df = pd.DataFrame(data)

# Extract mean and confidence intervals for plotting
df[['Mean', 'CI_Lower', 'CI_Upper']] = pd.DataFrame(df['Effect Size (mean, 95% CI)'].tolist(), index=df.index)
```

```python
# Forest Plot
plt.figure(figsize=(10, 8))
plt.errorbar(df['Mean'], df['Variable Subgroup'], 
             xerr=[df['Mean'] - df['CI_Lower'], df['CI_Upper'] - df['Mean']], 
             fmt='o', color='blue', ecolor='lightblue', elinewidth=3, capsize=5)
plt.axvline(x=0, color='grey', linestyle='--')
plt.title('Forest Plot of Effect Sizes by Variable Subgroup')
plt.xlabel('Effect Size (Mean)')
plt.ylabel('Variable Subgroup')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
# Bar Plot for Average Effect Sizes
plt.figure(figsize=(10, 6))
sns.barplot(x='Mean', y='Variable Subgroup', data=df, palette='viridis')
plt.title('Average Effect Sizes by Variable Subgroup')
plt.xlabel('Average Effect Size')
plt.ylabel('Variable Subgroup')
plt.grid(axis='x')
plt.tight_layout()
plt.show()

# Heatmap for I2 values
plt.figure(figsize=(10, 6))
```
